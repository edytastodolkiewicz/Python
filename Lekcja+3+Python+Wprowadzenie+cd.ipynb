# Instrukcje warunkowe

```python
# Instrukcje warunkowe if-elif-else
# if WARUNEK:
#   CO JEŚLI SPEŁNIONE
cena = 150
if cena >= 150:
    print("Zdecydowanie za drogo")
elif cena <= 150 and cena >= 100:
    print("Za drogo")
else:
    print("Kupuję")
```

```python
cena=100
if cena > 150:
    print("Zdecydowanie za drogo")
if cena <= 150 and cena >= 100:
    print("Za drogo")
if cena <= 100:
    print("Kupuję")
# MOGĄ BYĆ BLOKI: IF ELSE, IF ELIF, IF ELIF ELIF ELIF.... ELSE
# ZABRONIONE: ELSE, ELIF, ELIF ELSE
```

```python
# Instrukcja warunkowa w jednej linii
cena = 100
print("Zdecydowanie za drogo") if cena > 150 else print("Za drogo") if cena <= 150 and cena > 100 else print("Kupuję")
```

# Pętle w Pythonie

```python
# Suma liczb od 1 do 10
# for ITERATOR in ZAKRES:
#   WYKONANIE W PĘTLI

suma = 0
for i in range(1, 11):
    print(i)
    suma += i
print("\n", suma)
```

```python
# Suma liczb nieparzystych od -5 do 10
suma = 0
for i in range(-5, 11, 2):
    print(i)
    suma += i
print(suma)
```

```python
napis = "Ala ma kota"
for litera in napis:
    print(litera)
    #print("W pętli")
print("Poza pętlą")
```

```python
moja_lista = [{'imie': 'Adam'}, (1,2,3), ["ala", "kot"]]
for element in moja_lista:
    print(type(element), element)
    if type(element) == type({}):
        print(element['imie'])

#type(objekt)
moj_slownik = {} #[] lista, () krotka
moj_slownik['imie'] = 'Bartek'
print(moj_slownik)
```

```python
# Pętla while
# while WARUNEK:
#   PĘTLA
i = 10
while i > 0:
    print(i)
    i -= 1
else:
    print("Koniec")
```

```python
while True:
    print("Wirus!!!")
```

```python
# Break i continue w pętli
i = 10
while True:
    if i == 7:
        i = 6
        continue # KOŃCZY DANĄ ITERACJĘ
    if i == 2:
        break #CAŁKOWICIE WYCHODZI Z PĘTLI
    print(i)
    i = i - 1
else:
    print("Koniec")
    
print("Po pętli")
```

```python
# Funkcja zip
masa = [73, 89, 112, 54, 67]
wzrost = [175, 185, 178, 164, 172]

for x in zip(masa, wzrost):
    print(x)

for x, y in zip(masa, wzrost):
    print(x, y)
```

```python
# Obliczanie BMI
bmi = []
for _masa, _wzrost in zip(masa, wzrost):
    bmi.append(_masa/(_wzrost/100)**2)
print(bmi)
```

```python
# Wyrażenie listotwórcze
bmi2 = [_masa/(_wzrost/100)**2 for _masa, _wzrost in zip(masa, wzrost)]
print(bmi2)

moja_lista_2 = [1,2,3,4,5]
moja_lista_3 = [x**2 for x in moja_lista_2]
print(moja_lista_3)
```

# Funkcje w Pythonie

Funkcja w Pythonie (i nie tylko) jest to obiekt, który może byc wykonany i może zwrócic pewien rezultat.
W Pythonie, symbol wykonania to '()', umiejscowiony dokładnie po nazwie funkcji.
Funkcja może przyjmowac parametry (przekazywane wewnatrz '()' ) oddzielone przecinkami.

Parametry te widoczne sa w srodku funkcji tak jak zwykłe zmienne.

```python
#WAŻNE DEFINICJA FUNKCJI
'''
def nazwa_funkcji(argumenty):
  ciało_funkcji - wewnątrz funkcji
  ...
  return zwracane_wartości

nazwa_funkcji(przekazywane_argumenty)
'''

# Funkcja z jednym argumentem
def hello(name):
    print("Witaj", name+"!")
```

```python
hello("Andrzej")
hello("Miłosz")
```

```python
# Funkcja z dwoma argumentami (jeden z domyślną wartością), zwracająca wartość
def dodaj_vat(cena, stawka_vat=23):
    kwota_podatku = cena * (stawka_vat / 100)
    return cena + kwota_podatku

cena_brutto = dodaj_vat(150)
print(cena_brutto)
```

```python
def zwroc_text():
    return "Ala ma kota"

moj_tekst = zwroc_text()
print(moj_tekst)
```

```python
print(dodaj_vat(100))
print(dodaj_vat(cena=100, stawka_vat=15))
print(dodaj_vat(100, 23)) #trzeba zachować kolejność
print(dodaj_vat(stawka_vat=23, cena=100))
```

```python
print(dodaj_vat(100, stawka_vat=8))
print(dodaj_vat(1000))
```

```python
# Użycie * przed argumentem oznacza, że podajemy argumenty w postaci listy albo krotki - argsy
# Użycie ** przed argumentem oznacza, że podajemy argumenty w postaci słownika - kwargsy

# Rozpakowanie argumentów
a = (100, 8)
print(dodaj_vat(*a)) 
```

```python
# Lista argumentów o zmiennej długości
def wypiszWszystkie(*slowa):
    print(slowa)
    wynik = ""
    for x in slowa:
        wynik += x
        wynik += " "
    print(wynik)
```

```python
wypiszWszystkie("Witaj", "Andrzeju")
wypiszWszystkie("Dzień", "dobry", "Andrzeju")
wypiszWszystkie("No", "hej", "co", "tam", "słychać")
```

```python
# Lista argumentów nazwanych
def wypiszSlownik(**slowa):
    print(slowa)
    for klucz in slowa:
        wartosc = slowa[klucz]
        print(wartosc)    
```

```python
wypiszSlownik(marchew = 2.50, pomidor = 6.00, papryka = 11.99)
```

## Rekurencja

Rekurencja to przypadek, gdy funkcja wywołuje siebie sama.

```python
def rekurencyjna_suma(lista):
    suma = 0
    for element in lista:
        if isinstance(element, list):
            suma += rekurencyjna_suma(element)
        else:
            suma += element
    return suma
```

```python
print(rekurencyjna_suma([[1, 2, [3 ,7]], [4, 5], 6]))
```
